### Задание №2
Написать генератор случайных матриц(многомерных), который принимает
опциональные параметры <b>m</b>, <b>n</b>, <b>min_limit</b>, <b>max_limit</b>, где <b>m</b> и <b>n</b> указывают размер
матрицы, а <b>min_lim</b> и <b>max_lim</b> - минимальное и максимальное значение для
генерируемого числа.

```python
import random

user_m = int(input("Число строк матрицы:"))
user_n = int(input("Число стобцов матрицы:"))
user_min_limit = int(input("Минимальное значение для генерации:"))
user_max_limit = int(input("Максимальное значение для генерации:"))

def matrixGenerator(m, n, min_limit, max_limit):
    dmArray = []
    for i in range(m):
      newArray = []
      for j in range(n):
        newArray.append(random.randint(min_limit, max_limit))
      dmArray.append(newArray)
    
    return dmArray

matrix = matrixGenerator(user_m, user_n, user_min_limit, user_max_limit)
for row in matrix:
  print(row)
```

### Задание №3
Реализовать методы сортировки строк числовой матрицы в соответствии с
заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со
временем стандартной функции сортировки. Испытания проводить на сгенерированных
матрицах.

```python
import time
import copy
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.
start_time = time.time()
ourArray = copy.copy(matrix)

def selectionSort(ourArray):
  for i in range(len(ourArray)):
    minimumElement = i

    for j in range(i + 1, len(ourArray)):
      if ourArray[j] < ourArray[minimumElement]:
        minimumElement = j

    ourArray[minimumElement], ourArray[i] = ourArray[i], ourArray[minimumElement]
    return ourArray

for row in range(len(ourArray)):
  selectionSort(ourArray[row])

selectionSort(ourArray)

print("Сортировка выбором: ""--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка вставкой. Алгоритм делит список на две части, вставляя элементы на их правильные места 
# во вторую часть списка, убирая их из первой. Если второй элемент больше первого, то оставляем его на своем месте. 
# Если он меньше, то вставляем его на второе место, оставив первый элемент на первом месте. 
# Далее перемещаем большие элементы во второй части списка вверх, пока не встретим элемент меньше первого или не дойдем до конца списка.

start_time = time.time()

def insertionSort(ourArray):
  for i in range(len(ourArray)):
    temp = ourArray[i]
    position = i

    while position > 0 and ourArray[position - 1] > temp:
      # Меняем местами число, продвигая по списку
      ourArray[position] = ourArray[position - 1]
      position = position - 1
    # Останавливаемся и делаем последний обмен
    ourArray[position] = temp

  return ourArray

for row in range(len(ourArray)):
  insertionSort(ourArray[row])


#TODO
print("Сортировка вставкой: ""--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка обменом. || Сортировка пузырьком.
#Алгоритм попарно сравнивает элементы списка, меняя их местами, если это требуется. 
#Сравниваются первые два элемента. Если первый элемент больше, то они меняются местами. 
#Далее происходит все то же самое, но со следующими элементами до последней пары элементов в списке. 

start_time = time.time()

def bubbleSort(ourArray):
  changed = True
  while changed:
        changed = False
        for i in range(len(ourArray) - 1):
            if ourArray[i] > ourArray[i+1]:
                ourArray[i], ourArray[i+1] = ourArray[i+1], ourArray[i]
                changed = True
  return ourArray

for row in range(len(ourArray)):
     bubbleSort(ourArray[row]) 


print("Сортировка пузырьком: ""--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка Шелла.
#Шаг 0. i=t
#Шаг 1. Разобьем массив на списки элементов, отстающих друг от друга на Hi. Таких списков будет Hi
#Шаг 2. Отсортируем элементы каждого списка сортировкой вставками.
#Шаг 3. Объединим списки обратно в массив. Уменьшим i. Если i неотрицательно — вернемся к шагу 1

start_time = time.time()

def shellSort(ourArray):
    gap = len(ourArray) // 2 #Определяем шаг
    while gap > 0:
        for value in range(gap, len(ourArray)):
            currentValue = ourArray[value]
            position = value

            while position >= gap and ourArray[position - gap] > currentValue:
                ourArray[position] = ourArray[position - gap]
                position -= gap
                ourArray[position] = currentValue

        gap //= 2
    return ourArray

for row in range(len(ourArray)):
     shellSort(ourArray[row])


print("Сортировка Шелла: ""--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Быстрая сортировка.
# Сначала из массива нужно выбрать один элемент — его обычно называют опорным. Затем другие элементы в массиве перераспределяют так,
# чтобы элементы меньше опорного оказались до него, а большие или равные — после. А дальше рекурсивно применяют первые два шага 
# к подмассивам справа и слева от опорного значения.

start_time = time.time()

def quickSort(ourArray):
    if len(ourArray) <= 1:
      return ourArray

    greater = [] #Большие
    equal = [] #Равные
    lesser = [] #Меньше
    for i in range(len(ourArray)):
      if ourArray[i] == ourArray[0]:
        equal.append(ourArray[i])
      elif ourArray[i] > ourArray[0]:
        greater.append(ourArray[i])
      else:
        lesser.append(ourArray[i])
    return quickSort(lesser) + equal + quickSort(greater)


for row in range(len(ourArray)):
     quickSort(ourArray[row])

print("Быстрая сортировка:""--- {0} ms ---".format(round((time.time() - start_time)*1000)))
    
```

```python
# Турнирная сортировка.
#Преобразуем список в бинарное дерево, где самый большой элемент является вершиной дерева, и помещаем этот элемент в конец списка. 
#После перестраиваем дерево и помещаем новый наибольший элемент перед последним элементом в списке. 
#Повторяем этот алгоритм, пока все вершины дерева не будут удалены. 

start_time = time.time()

# Процедура для преобразования в двоичную кучу поддерева с корневым узлом rootIndex
def heapify(ourArray, heapSize, rootIndex):
  largest = rootIndex
  leftChild = 2 * rootIndex + 1
  rightChild = 2 * rootIndex + 2

  # Cуществует ли левый дочерний элемент больший, чем корень
  if leftChild < heapSize and ourArray[leftChild] > ourArray[largest]: 
    largest = leftChild

  # Cуществует ли правый дочерний элемент больший, чем корень
  if rightChild < heapSize and ourArray[rightChild] > ourArray[largest]:
    largest = rightChild

  # Заменяем корень, если нужно
  if largest != rootIndex:
    ourArray[rootIndex], ourArray[largest] = ourArray[largest], ourArray[rootIndex]
    heapify(ourArray, heapSize, largest)

def pyrSort(ourArray):
  length = len(ourArray)
  # Построение max-heap.
  for i in range(length, -1, -1):
    heapify(ourArray, length, i)
  # Один за другим извлекаем элементы
  for i in range(length - 1, 0, -1):
    ourArray[i], ourArray[0] = ourArray[0], ourArray[i]
    heapify(ourArray, i, 0)
  
  return ourArray

for row in range(len(ourArray)):
  pyrSort(ourArray[row])

print("Турнирная сортировка:""--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```
